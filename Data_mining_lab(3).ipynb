```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

iris = pd.read_csv('/content/Iris.csv')

X = iris.drop('Species', axis=1)
y = iris['Species']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X_scaled)
```

```python
reduced_df = pd.DataFrame(X_reduced, columns=['PC1', 'PC2']) # Change the columns to ['PC1', 'PC2']
reduced_df['Species'] = y

plt.figure(figsize=(8, 6))

plt.scatter(reduced_df['PC1'], reduced_df['PC2'], c=reduced_df['Species'].astype('category').cat.codes, cmap='viridis')

plt.xlabel('Principal Component 1 (PC1)')
plt.ylabel('Principal Component 2 (PC2)')
plt.title('PCA Dimensionality Reduction')
plt.colorbar(label='Species')
```
